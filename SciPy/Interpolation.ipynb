```python
import numpy as np
import matplotlib.pyplot as plt

X = np.arange(11)
Y = np.array([2.0, 1.9, 1.7, 1.5, 0.5, 0.0, 0.8, 2.0, 0.9, 0.2, 2.0])

plt.plot(X, Y, "o:")
plt.show()
```

```python
from scipy.interpolate import interp1d
predict = interp1d(X, Y, kind="quadratic")

X2 = np.linspace(0, 10, 100)
Y2 =np.array([predict(x) for x in X2])

plt.plot(X,Y ,"o:")
plt.plot(X2, Y2, "ro:")
plt.show()

print(f"X: 1.5, Y:{predict(1.5)}")
```

```python
from scipy.interpolate import interp1d
predict = interp1d(X, Y, kind="nearest")

X2 = np.linspace(0, 10, 100)
Y2 =np.array([predict(x) for x in X2])

plt.plot(X,Y ,"o:")
plt.plot(X2, Y2, "ro:")
plt.show()

print(f"X: 1.5, Y:{predict(1.5)}")
```

```python
from scipy.interpolate import interp1d
predict = interp1d(X, Y, kind="linear")

X2 = np.linspace(0, 10, 100)
Y2 =np.array([predict(x) for x in X2])

plt.plot(X,Y ,"o:")
plt.plot(X2, Y2, "ro:")
plt.show()

print(f"X: 1.5, Y:{predict(1.5)}")
```
